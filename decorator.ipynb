Global scope: changing the availability of the program making it accessible through ou every function

```python
num = 1000
def function():
  global num
  num = 100
  print(num)
function()
print(num)
```

Enclosed scope:the enclosed scope works with the concept of nested functional statements the outer function can take some arguments and inner function has access to all of them

```python
def outer(val1):
  def inner(val2):
    print(f'inner {val1 = }')
    print(f'inner {val2 = }')
  print(f'outer {val1 = }')
  return inner
fun = outer(9)
fun(19)
```

a decorater function which takes the other function as an input and modfy the code as per the decorator function and returns the value

```python
def absolute(fun):
  def updater(v1,v2):
    if v1>v2:
      return fun(v1,v2)
    else:
      return fun(v2,v2)
  return updater
def diff(num1,num2):
  return num1-num2
d1 = diff(10,5)
print(f'{d1 = }')
abso = absolute(diff)
d2 = abso(10,5)
print(f'{d2 = }')
```

Note:
the new modified function is not required to be stored in the different variable they can use the same
name in the function which is updated diff = absolute(diff)

With the help of decoration the storing is also not required the function which we are trying to modify 
gets teh address of the moderator function as a decorator Example:down

```python
@absolute
def diff(num1,num2):
  return num1-num2
```

after decoration even if we call the base function then the decorator function only wii be called

```python
d1 = diff(10,5)
print(f'{d1 = }')
d2 = diff(5,10)
print(f'{d2 = }')
```

When functions are having repitation we can use decorators to remove repetations but if we are having two diff types of dunctions which are taking varying length of positional arguments and keyword arguments and
same decorator is required to modify both functions we will use vla indicating varrying length of positional arguments and varying length of keyword arguments

```python
import time
def with_timer(fun):
  def updater(*args,**kwargs):
    start = time.time()
    fun(*args,**kwargs)
    end = time.time()
    print(end - start)
  return updater
@with_timer
def boost():
  print('making milk with boost..............')
@with_timer
def Tea(ind1):
  print(f'making Tea..........with {ind1}')
boost()
Tea(ind1 = 'Coco milk')
```
