```python
number = 20
prime = []
for i in range(2,number+1):
    for j in range(2,i):
        if i % j == 0:
            break
    else:
        prime.append(i)
print(prime)
```

```python
"""perfect number"""
num = 28
sum_od_divisors = 0
for i in range(1, num):
    if num % i == 0:
        sum_od_divisors += i
if sum_od_divisors == num:
    print(num, "is a perfect number")
else:
    print(num, "is not a perfect number") 
```

```python
"""diagonal star pattern 
"""
n=5
for row in range(1,n+1):
  for col in range(1,n+1):
    if row == col or col == n-row+1 or col == (n//2)+1:
      print('*',end=' ')
    else:
      print(' ',end=' ')
  print()
  
```

```python
n = 5
for row in range(1,n+1):
  for col in range(1,row+1):
    if col == 1 or row == n or row == col:
      print('*',end=' ')
    else:
      print(' ',end=' ')
  print()
```
